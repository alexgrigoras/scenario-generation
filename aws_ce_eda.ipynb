```python
import pandas as pd
```

```python
# Load datasets

input_path = '/Users/alexgrigoras/Datasets/aws-consumer-electronics/'
export_path = 'datasets/'

# Target time series:   https://github.com/aws-samples/amazon-forecast-samples/blob/main/library/content/TargetTimeSeries.md
data_tts = pd.read_csv(input_path + 'consumer_electronics_RTS.csv', parse_dates=True, index_col='timestamp')

#Related time series:  https://github.com/aws-samples/amazon-forecast-samples/blob/main/library/content/RelatedTimeSeries.md
data_rts = pd.read_csv(input_path + 'consumer_electronics_TTS.csv', parse_dates=True, index_col='timestamp')
```
